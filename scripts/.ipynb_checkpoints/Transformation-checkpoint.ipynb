```python
#drop the empty rows and columns
df_generalpaymentdata = df_generalpaymentdata.dropna(axis=1, how='all')
df_generalpaymentdata = df_generalpaymentdata.dropna(axis=0, how='all')
df_generalpaymentdata.info()
```

```python
#only keep the id needed
df_generalpaymentdata = df_generalpaymentdata[['Total_Amount_of_Payment_USDollars','Date_of_Payment',
                                              'Number_of_Payments_Included_in_Total_Amount',
                                              'Form_of_Payment_or_Transfer_of_Value',
                                              'Physician_Ownership_Indicator',
                                              'Program_Year',
                                              'Payment_Publication_Date',
                                              'Covered_Recipient_First_Name',
                                              'Covered_Recipient_Middle_Name',
                                              'Covered_Recipient_Last_Name',
                                              'Covered_Recipient_Primary_Type_1',
                                              'Covered_Recipient_Specialty_1',
                                              'Covered_Recipient_License_State_code1',
                                              'Nature_of_Payment_or_Transfer_of_Value',
                                              'Teaching_Hospital_ID',
                                              'Teaching_Hospital_Name',
                                              'Recipient_Primary_Business_Street_Address_Line1',
                                              'Recipient_City',
                                              'Recipient_State',
                                              'Recipient_Zip_Code']]
df_generalpaymentdata.info()
```

```python
#assign the data type
df_generalpaymentdata['Date_of_Payment'] = pd.to_datetime(df_generalpaymentdata['Date_of_Payment'])
df_generalpaymentdata['Payment_Publication_Date'] = pd.to_datetime(df_generalpaymentdata['Payment_Publication_Date'])

data_types = {
    'Total_Amount_of_Payment_USDollars': 'float',
    'Number_of_Payments_Included_in_Total_Amount': 'int',
    'Program_Year': 'int',
    'Teaching_Hospital_ID': 'string',
}

# Convert all 'object' type columns to 'string' type
for column, dtype in df_generalpaymentdata.dtypes.items():
    if dtype == 'object':
        data_types[column] = 'string'

# Apply the transformations
df_generalpaymentdata = df_generalpaymentdata.astype(data_types)

df_generalpaymentdata.info()
```

```python
# Creating Dimension Tables
# Dim Covered Recipient
dim_covered_recipient = df_generalpaymentdata[['Covered_Recipient_First_Name', 'Covered_Recipient_Middle_Name',
                            'Covered_Recipient_Last_Name', 'Covered_Recipient_Primary_Type_1',
                            'Covered_Recipient_Specialty_1', 'Covered_Recipient_License_State_code1']].drop_duplicates().reset_index(drop=True)
dim_covered_recipient['dim_Covered_Recipient_ID'] = range(1, len(dim_covered_recipient) + 1)

# Dim Teaching Hospital
dim_teaching_hospital = df_generalpaymentdata[['Teaching_Hospital_ID', 'Teaching_Hospital_Name',
                            'Recipient_Primary_Business_Street_Address_Line1', 'Recipient_City',
                            'Recipient_State', 'Recipient_Zip_Code']].drop_duplicates().reset_index(drop=True)
dim_teaching_hospital['dim_Teaching_Hospital_ID'] = range(1, len(dim_teaching_hospital) + 1)

# Dim Payment Detail
dim_payment_detail = df_generalpaymentdata[['Date_of_Payment', 'Program_Year', 'Payment_Publication_Date', 
                         'Form_of_Payment_or_Transfer_of_Value', 'Nature_of_Payment_or_Transfer_of_Value']].drop_duplicates().reset_index(drop=True)
dim_payment_detail['dim_Payment_Detail_ID'] = range(1, len(dim_payment_detail) + 1)

# Creating mappings
covered_recipient_mapping = dim_covered_recipient.set_index(['Covered_Recipient_First_Name', 'Covered_Recipient_Middle_Name',
                                                             'Covered_Recipient_Last_Name', 'Covered_Recipient_Primary_Type_1',
                                                             'Covered_Recipient_Specialty_1', 'Covered_Recipient_License_State_code1'])['dim_Covered_Recipient_ID']

teaching_hospital_mapping = dim_teaching_hospital.set_index(['Teaching_Hospital_ID', 'Teaching_Hospital_Name',
                                                             'Recipient_Primary_Business_Street_Address_Line1', 'Recipient_City',
                                                             'Recipient_State', 'Recipient_Zip_Code'])['dim_Teaching_Hospital_ID']

payment_detail_mapping = dim_payment_detail.set_index(['Date_of_Payment', 'Program_Year', 'Payment_Publication_Date', 
                                                       'Form_of_Payment_or_Transfer_of_Value', 'Nature_of_Payment_or_Transfer_of_Value'])['dim_Payment_Detail_ID']

# Map IDs to the Fact Table
df_generalpaymentdata['dim_Covered_Recipient_ID'] = df_generalpaymentdata.set_index(['Covered_Recipient_First_Name', 'Covered_Recipient_Middle_Name',
                                               'Covered_Recipient_Last_Name', 'Covered_Recipient_Primary_Type_1',
                                               'Covered_Recipient_Specialty_1', 'Covered_Recipient_License_State_code1']).index.map(covered_recipient_mapping)

df_generalpaymentdata['dim_Teaching_Hospital_ID'] = df_generalpaymentdata.set_index(['Teaching_Hospital_ID', 'Teaching_Hospital_Name',
                                               'Recipient_Primary_Business_Street_Address_Line1', 'Recipient_City',
                                               'Recipient_State', 'Recipient_Zip_Code']).index.map(teaching_hospital_mapping)

df_generalpaymentdata['dim_Payment_Detail_ID'] = df_generalpaymentdata.set_index(['Date_of_Payment', 'Program_Year', 'Payment_Publication_Date', 
                                            'Form_of_Payment_or_Transfer_of_Value', 'Nature_of_Payment_or_Transfer_of_Value']).index.map(payment_detail_mapping)

# Fact General Payments Table
fact_general_payments = df_generalpaymentdata[['Total_Amount_of_Payment_USDollars', 
                 'Number_of_Payments_Included_in_Total_Amount', 
                 'Physician_Ownership_Indicator',
                 'dim_Covered_Recipient_ID',
                 'dim_Teaching_Hospital_ID',
                 'dim_Payment_Detail_ID']].copy()
fact_general_payments['Fact_id'] = range(1, len(fact_general_payments) + 1)

print(dim_covered_recipient.head())
print(dim_teaching_hospital.head())
print(dim_payment_detail.head())
print(fact_general_payments.head())
```
